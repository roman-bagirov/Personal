```python
for i in df_train.columns:                      # перебираем все столбцы
    if str(df_train[i].dtype) == 'float64':      # если тип столбца - object
        print('\n') 
        print(i)                                # выводим название столбца
        print(set(df_train[i]))                 # выводим все его значения (но делаем set - чтоб значения не повторялись)
        print('\n') 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
X = df_train.loc[:, (df_train.columns != 'use_direct_debit') & (df_train.columns != 'customer_id')]
y = df_train.loc[:, df_train.columns == 'use_direct_debit']
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y, test_size=0.3, random_state=42)
```

```python
Decision_tree = DecisionTreeClassifier(criterion  = 'gini', min_samples_leaf = 5, min_samples_split = 10)
Decision_tree.fit(X_train, y_train)
```

```python
Decision_tree.feature_importances_
```

```python
Prediction_test = Decision_tree.predict(X_test)
```

```python
f1_validation = f1_score(Prediction_test, y_test)
print(f1_validation)
```

```python
df_test_data = df_test_data.drop('customer_id', axis = 1)
```

```python
kaggle_output = Decision_tree.predict(df_test_data)

kaggle_DT = pd.DataFrame(zip(df_test_sample['customer_id'],kaggle_output), columns = ['customer_id', 'use_direct_debit'])
```

```python
for i in range(2,40):
    for j in range(2,40):
        for h in range(2,40):
                for z in range(2,26):
                    Decision_tree = DecisionTreeClassifier(criterion  = 'gini', 
                                min_samples_leaf = i, min_samples_split = j, max_depth = h, max_features = z)
                    Decision_tree.fit(X_train, y_train)
                    f1_value = {'F1' : f1_score(Decision_tree.predict(X_test), y_test),
                                'min_samples_leaf' : i,'min_samples_split' : j,'max_depth' : h, 'max_features': z
                               } 
                    df = df.append(f1_value, ignore_index=True)
```
